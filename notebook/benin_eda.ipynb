```python
import pandas as pd

# Reload Benin data, skip the second header row
benin = pd.read_csv('../data/solar-measurements_benin-malanville_qc.csv',
                    encoding='latin1', 
                    skiprows=[1], 
                    low_memory=False)

# Check first few rows
benin.head()


```

```python
# --- 1. Inspect basic info ---
benin.info()
print("\nMissing values per column:")
print(benin.isna().sum())

# --- 2. Convert timestamp to datetime ---
benin['Timestamp'] = pd.to_datetime(benin['Timestamp'], errors='coerce')

# --- 3. Drop rows with no valid solar measurements ---
benin = benin.dropna(subset=['GHI', 'DNI', 'DHI'], how='all')

# --- 4. Remove impossible (negative) values ---
for col in ['GHI', 'DNI', 'DHI']:
    benin.loc[benin[col] < 0, col] = None

# --- 5. Handle missing timestamps (optional) ---
benin = benin.sort_values('Timestamp').reset_index(drop=True)

# --- 6. Save cleaned dataset ---
benin.to_csv('../data/benin_clean.csv', index=False)
print("‚úÖ Benin dataset cleaned and saved successfully!")
```
